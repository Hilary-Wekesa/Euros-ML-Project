```python
fixtures_url="https://fbref.com/en/comps/676/schedule/UEFA-Euro-Scores-and-Fixtures"
```

```python
years=[2024, 2021]

import time

all_matches=[]
fixtures_url="https://fbref.com/en/comps/676/schedule/UEFA-Euro-Scores-and-Fixtures"

for year in years:
    data=requests.get(fixtures_url)
    soup=BeautifulSoup(data.text)
    
    fixtures_table=soup.select('table.stats_table')[0]
    
    links=fixtures_table.find_all('a')
    links=[l.get("href") for l in links]
    links=[l for l in links if '/squads/' in l]
    
    filtered_links=[]
    filtered_links=list(dict.fromkeys(links))
    
    team_urls = [f"https://fbref.com{l}" for l in filtered_links]
    
    previous_season = soup.select("a.prev")[0].get("href")
    fixtures_url = f"https://fbref.com{previous_season}"
    
    for team_url in team_urls:
        team_name = team_url.split("/")[-1].replace("-Stats", "").split("-")[0]
        data=requests.get(team_url)
        team_matches=pd.read_html(data.text,match="Scores & Fixtures ")
        team_data=team_matches[0]
        time.sleep(2)
        
        soup_2=BeautifulSoup(data.text)
        stat_links=soup_2.find_all('a')
        stat_links=[l.get("href") for l in stat_links]
        
        shooting_link=[l for l in stat_links if l and 'all_comps/shooting' in l][0]
        shooting_data=requests.get (f"https://fbref.com{shooting_link}")
        shooting=pd.read_html(shooting_data.text,match="Shooting")[0]
        shooting.columns=shooting.columns.droplevel()
        time.sleep(2)
        
        gca_link=[l for l in stat_links if l and 'all_comps/gca' in l][0]
        gca_data=requests.get(f"https://fbref.com{gca_link}")
        gca=pd.read_html(gca_data.text,match="Goal and Shot Creation")[0]
        gca.columns=gca.columns.droplevel()
        time.sleep(2)
        
        poss_link=[l for l in stat_links if l and 'all_comps/possession' in l][0]
        poss_data=requests.get(f"https://fbref.com{poss_link}")
        poss=pd.read_html(poss_data.text,match="Possession")[0]
        poss.columns=poss.columns.droplevel()
        time.sleep(2)
        
        try:
            team_data=team_data.merge(shooting[['Date','Sh', 'SoT', 'SoT%', 'G/Sh', 'G/SoT', 'Dist','PK', 'PKatt','G-xG', 'np:G-xG']],on="Date")
            team_data=team_data.merge(gca[['Date','SCA','GCA']],on="Date")
            team_data=team_data.merge(poss[['Date','Att 3rd', 'Att Pen','Att', 'Succ', 'Succ%','PrgC', '1/3', 'CPA']],on="Date")
        except ValueError:
            continue
            
        team_data["Year"] = year
        team_data["Team"] = team_name
        all_matches.append(team_data)
        print(f"{team_name} in {year} added")
        time.sleep(5)
        
        
```

```python
match_df = pd.concat(all_matches)
match_df.columns = [c.lower() for c in match_df.columns]
match_df.to_csv("euros_matches.csv")
```
